## Descripción

Una caminata aleatoria 1D es un modelo donde una partícula se mueve en una línea recta.

En cada paso:
 Se mueve a la derecha con probabilidad $p$

Se mueve a la izquierda con probabilidad $q$

cumpliendo:

$$
p + q = 1
$$

Si llamamos $X_n$ a la posición después de $n$ pasos:

$$
X_n = \sum_{i=1}^{n} S_i
$$

donde:

$$
S_i =
\begin{cases}
+1 & \text{con probabilidad } p \\
-1 & \text{con probabilidad } q
\end{cases}
$$



```python
import numpy as np
import matplotlib.pyplot as plt
import random as r
```

```python
n_pasos = 100
p = 0.7
# guardar posiciones
posiciones = [0]
# caminata
for i in range(n_pasos):
  r = np.random.rand()  # numero aleatorio entre 0 y 1
  if r < p:
    paso = 1
  else:
    paso = -1
  posiciones.append(posiciones[-1] + paso)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(posiciones)
plt.title('Caminata Aleatoria')
plt.xlabel('Pasos')
plt.ylabel('Posición')
plt.grid(True)
plt.show()
```

```python
print(posiciones)
```

```python
print("Posición final:", posiciones[-1])
```

```python
promedio = np.mean(posiciones)
print("Posición promedio:", promedio)

#Si p = 0.5 promedio cerca de 0
#Si p > 0.5 promedio positivo
#Si p < 0.5 promedio negativo
```
