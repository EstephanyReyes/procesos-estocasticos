#Tutorial para multiplicar matrices, escribir una matriz y multiplicar matriz por vector

Para trabajar con matrices en SymPy, se siguen los sientes pasos:

Paso 1. Importar la librería

Paso 2. Se definen utilizando la clase $Matrix$, pasándole como argumentos una lista de listas, donde cada sublista corresponde a una fila de la matriz
$$Matrix([[fila1], [fila2], ...])$$




```python
import sympy as sp
from IPython.display import display
```

```python

# 1. Representación de matrices
A = sp.Matrix([[1, 2],
               [3, 4]])

B = sp.Matrix([[1, 0, 0],
               [0, 1, 0],
               [0, 0, 1]])

print("Representación de matrices")
print("Matriz A:")
display(A)
print("Matriz B:")
display(B)
```

**Obtener una entrada específica, renglón o columna**

En SymPy los índices empiezan en 0.

Para una entrada específica:
$$A[i, j]$$ esto es el elemento en fila i , columna j

Obtener un renglón:
$$A.row(i)$$

Obtener una columna
$$A.col(j)$$

```python
A = sp.Matrix([[1, 2],
               [3, 4]])

display(A)
display(B)
print("Entrada específica")
print("A[1,1]:", A[1,1])
print("Obtener un renglón")
display(A.row(0))
print("Obtener una columna")
display(A.col(0))
```

**Multiplicar matrices o matriz por vector**

Para el producto matricial estándar (fila por columna) se utiliza el operador asterisco $"*"$.

La matriz A debe tener tantas columnas como filas tiene B.

Un vector se define como una matriz con una sola columna:
$$Matrix([x, y, z])$$
Un vector fila de define como una lista dentro de otra lista:
$$Matrix([[x, y, z]])$$


```python
# 2. Multiplicar matrices (A * B)
A = sp.Matrix([[1, 2],
               [3, 4]])

B = sp.Matrix([[5, 6],
               [7, 8]])

resultado_mat = A * B

print("Multiplicación de matrices")
print("Resultado A*B:")
display(resultado_mat)

# 3. Multiplicar matriz por vector

v = sp.Matrix([1, 2]) # vector columna
producto_mat_vec_colum = A * v

w = sp.Matrix([[1, 2]]) # vector fila
producto_mat_vec_fila = w * A

print("Multiplicar matriz por vector")
print("Resultado A*v:\n")
display(producto_mat_vec_colum)
print("Resultado w*A:\n")
display(producto_mat_vec_fila)

```


**Elevar matrices a una potencia**

Para elevar una matriz a una potencia, se utiliza el operador doble asterisco $"**"$ sobre un objeto $Matrix$. La matriz debe ser cuadrada (mismo número de filas y columnas).

```python


# 4. Elevar matriz a una potencia

# Definir la matriz
A = sp.Matrix([[1, 2], [3, 4]])

# Elevar al cuadrado (A^2)
A_cuadrado = A**2

print("Elevar matriz a una potencia")
print("Matriz original:")
display(A)
print("\nMatriz al cuadrado:")
display(A_cuadrado)
```

**Valores y vectores propios**

Para obtener los valores propios (eigenvalues) utilizamos: $$A.eigenvals()$$

y para los vectores propios (eigenvectors) utilizamos:  $$A.eigenvects()$$

Nos devolverán una lista de la forma:

(valor_propio, multiplicidad, [lista_de_vectores_propios])

```python

# 5. Valores y vectores propios
A = sp.Matrix([[2, 1],
               [1, 2]])

# Valores propios
A.eigenvals()

# Vectores propios
A.eigenvects()


for val, mult, vects in A.eigenvects():
    print("Valor propio:", val)
    print("Multiplicidad:", mult)
    print("Vector propio:")
    display(vects[0])
    print()
```

#Dibujar un diagrama de transición con NetworkX
Un diagrama de transición (por ejemplo en cadenas de Markov) es un grafo dirigido donde:


*   Cada nodo representa un estado
*  Cada flecha representa una probabilidad de transición






Paso 1: Importar librerias

```python
import networkx as nx
import matplotlib.pyplot as plt
```

Paso 2: crear un grafo dirigido

```python
G = nx.DiGraph()
```

Paso 3: Agregar estados (nodos)

```python
G.add_nodes_from(["A", "B"])
```

Paso 4: Agregar transiciones (aristas con peso)

Supongamos la matriz de transición:

P = \begin{bmatrix}
    {0.7} & {0.3} \\
    {0.4} & {0.6}
\end{bmatrix}

Esto significa:

A → A con 0.7

A → B con 0.3

B → A con 0.4

B → B con 0.6

Se agregan así:

```python
G.add_edge("A", "A", weight=0.7)
G.add_edge("A", "B", weight=0.3)
G.add_edge("B", "A", weight=0.4)
G.add_edge("B", "B", weight=0.6)
```

Paso 5: Dibujar el diagrama

```python
# Posiciones fijas
pos = {"A": (-1, 0), "B": (1, 0)}

# Dibujar nodos
nx.draw_networkx_nodes(G, pos, node_size=500, node_color="lightgreen")
nx.draw_networkx_labels(G, pos)


# Dibujar aristas con curva
nx.draw_networkx_edges(
    G, pos,
    connectionstyle='arc3,rad=0.5',  # curva las aristas
    arrows=True,
    arrowstyle='-|>',
    arrowsize=23
)

# Etiquetas
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    label_pos=0.9, #posicion de las etiquetas
    rotate= False
)

plt.axis('off')
plt.show()
```
